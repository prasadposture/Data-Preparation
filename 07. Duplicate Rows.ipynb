# Duplicate Rows
In here we will see how to get the duplicate rows i.e. the rows having all the column entries simliar and deal with them.

```python
#Importing libraries
import pandas as pd
```

```python
#Loading the data
cols=['user_id','Age','gender','occupation','zip_code']
df=pd.read_csv('http://bit.ly/movieusers','|',names=cols,index_col='user_id')
df.head()
```

```python
df.shape
```

If we want to see whether the particular column of our dataframe contains duplicate entry:

```python
df.zip_code.duplicated()
```

```python
df.zip_code.duplicated().sum()
#gives sum
```

If we want to identify whether the whole row is duplicated or not we use `.dupplicated()` which returns boolean value.

```python
df.duplicated()
```

```python
df.duplicated().sum()
#gives sum
```

```python
#Getting all the rows which are duplicated
df.loc[df.duplicated()]
```

```python
#Getting the first rows of the duplicate rows
df.loc[df.duplicated(keep='first')]
```

```python
#Getting the last rows of the duplicate rows
df.loc[df.duplicated(keep='last')]
```

```python
#Getting all duplicated rows
df.loc[df.duplicated(keep=False)]
```

**Dropping Dupicate**

```python
df.drop_duplicates(keep='first').shape
#first 7
#Gives shape if we drop the first rows of the duplicated rows
#df.drop_duplicates(keep='first', inplace=True) for dropping
```

```python
df.drop_duplicates(keep='last').shape
#last 7
#Gives shape if we drop the last rows of the duplicated rows
```

```python
#drop all 14 rows
df.drop_duplicates(keep=False).shape
#gives shape
```

```python
#only consider age and zip_code as relevant columns for duplication
df.duplicated(subset=['Age','zip_code']).sum()
```

```python
df.loc[df.duplicated(subset=['Age','zip_code'])]
#by deafult it gives the first rows of the duplicate rows
```

```python
df.loc[df.duplicated(subset=['Age','zip_code'],keep=False)]
#gives all the rows
```
